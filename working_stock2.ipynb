```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
from statsmodels.tsa.arima.model import ARIMA
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# --- 1. Fetch price data ---
def fetch_polygon_data(symbol, days=600):
    from datetime import datetime, timedelta
    API_KEY = 'n2J3phB6PlBhEG_jv5VndSs3o3iopbs7'
    end_date = datetime.now()
    start_date = end_date - timedelta(days=days)
    url = f"https://api.polygon.io/v2/aggs/ticker/{symbol}/range/1/day/{start_date:%Y-%m-%d}/{end_date:%Y-%m-%d}?adjusted=true&sort=asc&limit=5000&apiKey={API_KEY}"
    resp = requests.get(url)
    data = resp.json()
    if 'results' in data:
        df = pd.DataFrame(data['results'])
        df['t'] = pd.to_datetime(df['t'], unit='ms')
        df.rename(columns={'c': 'Close', 't': 'Date'}, inplace=True)
        return df[['Date', 'Close']]
    else:
        print(f"Error fetching {symbol}: {data}")
        return pd.DataFrame()

symbol = 'GOOG'
price_data = fetch_polygon_data(symbol)

# --- 2. Fetch news and compute sentiment ---
ALPHA_VANTAGE_API_KEY = 'G8ZRQWUYYRPCVAPO'
url = f'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={symbol}&apikey={ALPHA_VANTAGE_API_KEY}'
response = requests.get(url)
data = response.json()

headlines = []
if 'feed' in data:
    for item in data['feed']:
        headlines.append(item['title'])
else:
    print("No news found or API limit reached.")

tokenizer = AutoTokenizer.from_pretrained('ProsusAI/finbert')
finbert = AutoModelForSequenceClassification.from_pretrained('ProsusAI/finbert')

def get_finbert_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        outputs = finbert(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)
        sentiment = ['negative', 'neutral', 'positive'][probs.argmax()]
        score = probs.max().item()
    return sentiment, score

results = []
for headline in headlines:
    sentiment, score = get_finbert_sentiment(headline)
    results.append({'headline': headline, 'sentiment': sentiment, 'confidence': score})

sentiment_map = {'positive': 1, 'neutral': 0, 'negative': -1}
sentiment_df = pd.DataFrame(results)
sentiment_df['sentiment_score'] = sentiment_df['sentiment'].map(sentiment_map)

if 'time_published' in data['feed'][0]:
    sentiment_df['Date'] = pd.to_datetime(
        [item['time_published'][:8] for item in data['feed']],
        format='%Y%m%d'
    )
else:
    sentiment_df['Date'] = pd.Timestamp.today().normalize()

sentiment_data = sentiment_df.groupby('Date')['sentiment_score'].mean().reset_index()

# --- 3. Merge price and sentiment ---
merged_df = pd.merge(price_data, sentiment_data, on='Date', how='left')
merged_df['sentiment_score'].fillna(0, inplace=True)
merged_df = merged_df.sort_values('Date').reset_index(drop=True)

# --- 4. Train/test split ---
split_idx = int(len(merged_df) * 0.8)
train_df = merged_df.iloc[:split_idx]
test_df = merged_df.iloc[split_idx:]

# --- 5. ARIMA model on training price ---
train_close = train_df['Close'].values
arima_model = ARIMA(train_close, order=(5,1,0)).fit()
arima_pred_train = arima_model.predict(start=1, end=len(train_close)-1, typ='levels')

# --- 6. Compute ARIMA residuals ---
residuals = train_close[1:] - arima_pred_train

# --- 7. Prepare LSTM data on residuals (optionally add sentiment) ---
window = 100
# For sentiment, align with residuals (skip first value)
train_sentiment = train_df['sentiment_score'].values[1:]

X_lstm, y_lstm = [], []
for i in range(window, len(residuals)):
    # Use both residuals and sentiment as features
    seq_residuals = residuals[i-window:i].reshape(-1, 1)
    seq_sentiment = train_sentiment[i-window:i].reshape(-1, 1)
    X_lstm.append(np.hstack([seq_residuals, seq_sentiment]))
    y_lstm.append(residuals[i])
X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)

# --- 8. Build and train LSTM on residuals ---
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(window, 2)))
model.add(Dropout(0.2))
model.add(LSTM(60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(120, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_lstm, y_lstm, epochs=10, batch_size=32, verbose=1)

# --- 9. Hybrid prediction for next 7 days ---
future_days = 7
arima_forecast = arima_model.forecast(steps=future_days)

# For LSTM, use last window of residuals and sentiment
last_residuals = residuals[-window:]
last_sentiments = train_sentiment[-window:]
X_pred = np.hstack([last_residuals.reshape(-1, 1), last_sentiments.reshape(-1, 1)]).reshape(1, window, 2)

lstm_forecast = []
for _ in range(future_days):
    res_pred = model.predict(X_pred, verbose=0)[0,0]
    lstm_forecast.append(res_pred)
    # Update input for next prediction
    next_sentiment = 0  # or use last known sentiment
    next_row = np.array([[res_pred, next_sentiment]])
    X_pred = np.append(X_pred[:,1:,:], next_row.reshape(1,1,2), axis=1)

# --- 10. Combine ARIMA and LSTM predictions ---
final_forecast = arima_forecast + np.array(lstm_forecast)

# --- 11. Plot ---
last_date = merged_df['Date'].iloc[-1]
future_dates = pd.date_range(last_date, periods=future_days+1, inclusive='right')

plt.figure(figsize=(10,6))
plt.plot(merged_df['Date'][-50:], merged_df['Close'][-50:], label='Historical Price')
plt.plot(future_dates[1:1+len(final_forecast)], final_forecast[:len(future_dates[1:1+len(final_forecast)])], 'ro-', label='Hybrid ARIMA+LSTM Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Next 7 Days Stock Price Prediction (ARIMA + LSTM)')
plt.legend()
plt.show()

print("Hybrid ARIMA+LSTM forecast:", final_forecast)
```
